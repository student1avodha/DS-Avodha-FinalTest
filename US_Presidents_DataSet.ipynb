```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("us_presidents.csv")

data
```

```python
data.shape
```

```python
data.info()
```

```python
# drop unnamed feature

data.drop(columns=['Unnamed: 0'],inplace=True)

data.columns
```

```python
# rename feature 'S.No'


data.rename(columns={'S.No.':'President No'},inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=[14,8])

plt.hist(data['presidency_begin_age'])

plt.title('Presidency begin Age',fontname='Times new roman',fontsize=17)

plt.ylabel('Frequency')


plt.xlabel('presidency_begin_age')

plt.show()
```

```python
plt.figure(figsize=[20,10])

base_color=sns.color_palette()[0]

order=data['party'].value_counts().index

sns.countplot(data=data,x='party',order=order,color=base_color);

plt.title('Parties of US Presidents',fontsize=19,fontname='Times new roman');

plt.xticks(rotation=90)
```
