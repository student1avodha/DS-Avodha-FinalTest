#### EDA

Which season had most number of matches?

```python
fig_dims =(25,15)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Season',ax = ax , data=data)
plt.show
```

which is the most succesful IPL team with all the data at hand?

```python
plt.figure(figsize=(25,20))

data1 = data.winner.value_counts()
sns.barplot(y = data1.index , x = data1)
```

what is the probability of winning a match if the toss was won?


```python
prob_of_win = data['toss_winner'] == data['winner']

prob_of_win.groupby(prob_of_win).size()
```

```python
sns.countplot(prob_of_win)
```

Highest wins by teams per season

```python
data.groupby('Season')['winner'].value_counts()
```

```python
# plotting toss decision

sns.countplot(x=data['toss_decision'])
```

```python
data['toss_decision'].value_counts()
```

Man of the match - Highest to lowest top 10(in won matches)

```python
data['player_of_match'].value_counts().head(10)
```

```python
#visualize highest number of times man of the match getting players  ( Top ten)

plt.figure(figsize=(25,15))

sns.barplot(x = data['player_of_match'].value_counts().head(10).index , y =data['player_of_match'].value_counts().head(10) )
```

```python
data.player_of_match.fillna(value='data not given',inplace=True)
```

```python
from wordcloud import WordCloud
```

```python
# ploting the wordcloud for the highest number of times man of the match getting players  


plt.subplots(figsize=(25,15))



wordcloud = WordCloud(

                         background_color = 'white',
                          
                         width = 1920,
    
                         height = 1000

                       ) . generate(" ".join(data.player_of_match))




plt.imshow(wordcloud)


plt.axis('off')

#plt.savefig('graph.png')


plt.show()
```

Team Performance

```python
data1=data[data['win_by_runs']>0].groupby(['winner'])['win_by_runs'].apply(np.median).sort_values(ascending = False)

data1
```

```python
plt.figure(figsize=(15,12))
#fig, ax = plt.subplots()


ax.set_title("Winning by Runs - Team Performance")

sns.boxplot(y = 'winner', x = 'win_by_runs', data=data[data['win_by_runs']>0], orient = 'h'); #palette="Blues");
plt.show()
```

In which city were the number of matches played?

```python
data['city'].value_counts()
```

```python
#visualize most number of match held venue  ( Top ten)

plt.figure(figsize=(25,15))

sns.barplot(x = data['city'].value_counts().head(10).index , y =data['city'].value_counts().head(10) )

plt.show ()
```

```python
data.city.fillna(value='data not given',inplace=True)
```

```python
# ploting the wordcloud for the highest number of matches held venues  


plt.subplots(figsize=(20,15))



wordcloud = WordCloud(

                         background_color = 'black',
                          
                         width = 1920,
    
                         height = 1000

                       ) . generate(" ".join(data.city))




plt.imshow(wordcloud)


plt.axis('off')

#plt.savefig('graph.png')


plt.show()
```
